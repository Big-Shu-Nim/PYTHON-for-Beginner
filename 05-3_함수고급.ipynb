# Map () Function 

Map(function, list):
리스트의 element를 함수에 넣고 리턴된 값으로 새로운 리스트를 구성해주는 함수 


```python
# 함수 선언

def power(number):
    return number * number
def under_3(item):
    return item < 3

# 변수 선언

list_input_a = [1, 2, 3, 4, 5]

# map 함수 

output_a = map(power, list_input_a)
print("map함수 사용 결과 출력")
print("map(power, list)input_a):", output_a)  # 함수선언시 object에 담기게 된다
print("map(power, list)input_a):", list(output_a)) # 따라서 list()를 사용하여 강제 호출해야한다.
```

##  Filter() Function
filter(function, list): 리스트의 요소를 함수에 넣고 리턴된 값이 True인것으로 새로운 리스트를 구성하는 함수



```python
output_b = filter(under_3, list_input_a)
print("filter 사용 결과 출력")
print("filter(under_3, list)input_a):", output_b)  # 함수선언시 object에 담기게 된다 이를 generator라고 한다
print("filter(under_3, list)input_a):", list(output_b)) # 따라서 list()를 사용하여 강제 호출해야한다.

```

## lambda
매번 argumnet를 작성하여 함수내 변수에 parameter을 전달하는게 번거롭고, 코드 공간 낭비라는 생기 들수 있습니다.

lambda argument: return 
power()로 에르들자면 def키워드로 선언했던 함수를 lambda로 바꾸고 return 키워드를 따로 쓰지 않을수 있는 차이를 만들어 냈습니다. 


```python

power = lambda x: x*x
under_3 = lambda x: x<3

# 변수 선언

list_input_a = [1, 2, 3, 4, 5]

# map 함수 

output_a = map(power, list_input_a)
print("map함수 사용 결과 출력")
print("map(power, list)input_a):", output_a)  # 함수선언시 object에 담기게 된다
print("map(power, list)input_a):", list(output_a)) # 따라서 list()를 사용하여 강제 호출해야한다.


# filter
output_b = filter(under_3, list_input_a)
print("filter 사용 결과 출력")
print("filter(under_3, list)input_a):", output_b)  # 함수선언시 object에 담기게 된다 이를 generator라고 한다
print("filter(under_3, list)input_a):", list(output_b)) # 따라서 list()를 사용하여 강제 호출해야한다.


```

## 텍스트 한 줄 씩 읽기

    수십만명의 데이터를 저장하고 활용하는경우가 많습니다.
    하지만 모든 데이터를 올려놓고 사용하는 것은 컴퓨터의 성능에 악역향을 끼칠 수 있습니다.
    도마 위에 야채를 한번에 모두 올려놓고 자르지 않듯이, 한번에 자를수 있는 양만큼 오려서 자르고 처리하는것을 반복하죠.
    데이터도 마찬가지 입니다.
    
    간단하게 이에 대한 실습을 해봅시다.

```python
# 랜덤하게 1000명의 키와 몸무게 만들기

import random

hanguls = list('가나다라마바사아자차카타파하')

#쓰기모드로 염
with open('source/info.txt', 'w') as file:
    for i in range(1000):
        # 랜덤변수 생선
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40, 100)
        height= random.randrange(140, 200)
        
        #텍스트를 씁니다.
        file. write("{}, {}, {} \n".format(name, weight, height))
```

```python
# 반복문으로 파일 한 줄 씩 읽기

with open("source/info.txt", "r") as file:
    for line in file:
        # 변수 선언
        (name, wieght, height) = line.strip().split(", ")
        
        #데이터가 문제없는지 확인합니다. 문제가 있으면 지나감
        
        if (not name) or (not weight) or (not height):
            continue
        # calcuate the result
        
        bmi = int(weight) / ((int(height) / 100) **2)
        result = ""
        if 25 <= bmi:
            result  = 'overweight'
        elif 18.5 <= bmi :
            result = 'normal'
        else:
            result = 'underweight'
        
        # print out
        
        print('\n'.join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "BMI: {}",
            "결과: {}"
        ]).format(name, weight, height, bmi, result))
        print()
```

## 제너레이터, 이터레이터, yield, next()

Generattor는 이터레이터를 직접 만들때 사용하는 코드입니다.

함수 내부에 yield 키워드를 사용하면 해당 함수는 제너레이터 함수가 되면 일반 함수와 달리 함수를

호출해도 실행되지 않습니다.



```python
# generator function

def test():
    print('a function is called')
    yield 'test'

test()
```

```python
# 재너레이터 객체와 next함수

# 함수를 선언합니다.

def test():
    print("passed the point A")
    yield 1
    print('passed the point B')
    yield 2
    print('passed the point c')
# a function is called
output = test()

# next() function is called

print('passed the point D')
a = next(output)
print(a)
print('Passed the point E')
b = next(output)
print('Passed the point F')

c = next(output)
print(c)
next(output)
```

next 함수를 호출한 이후 yield키워드를 만나지 못하면 stopriteration 발생


# reversed()함수로 리스트 뒤집기
함수의 알규먼트에 리스트를 넣으면 리스트를 뒤집을수 있습니다.


```python
#reversed() function

list_a = [1,2,3,4,5]
list_reversed = reversed(list_a)
print(list_reversed)
print(list(list_reversed))

for i in reversed(list_a):  # 반복문 적용 따로 함수결과 활용하지 않고 for 구문 내부에서 사용됨 
    print(i)
```

```python
ex = reversed([1,2,3,4,5,6])

for i in ex:
    print(' 첫번째 반복문 {}'.format(i))
for i in ex:
    print(' 두번째 반복문 {}'.format(i))

# 이처럼 함수를 선언하면 "제너레이터"이기에 두번째 부분은 나오지 않는다. 

```

```python
# 필요할때 직접 for 구문에 넣어서 사용해야한다

for i in reversed([1,2,3,4,5,6]):
    print(' 첫번째 반복문 {}'.format(i))
for i in reversed([1,2,3,4,5,6]):
    print(' 두번째 반복문 {}'.format(i))



```

#  리스트 내포 list comprehensions

반복문을 활용해 리스트를 재조합 하는 경우가 많다.
예르들어 다음 코드에서는 짝수를 구한뒤 제곱해서 새로운 리스트를 만든다. 



```python
# 반복을 사용한 리스트 생성

array = []

for i in range(0 , 20, 2):
    array.append(i * i)
print(array)
```

이런 결과를 파이썬은 한줄로 줄일 수 있다.


```python
array = [i * i for i in range(0, 20, 2)]
print(array)
```
